#Short script to help you get on-chain historical transactional data using Covalent API ðŸ˜ƒ

 
Covalent is an Unified APi, it contains a vast option of endpoints for various networks.  

You will need to register to get an API-KEY , you can do it here: 
- https://www.covalenthq.com/  

The script is configured to get historical data, its using the endpoint `'transactions_v2'` and other configurations on the url : `'?quote-currency=USD&format=JSON&block-signed-at-asc=false&no-logs=true'`. If you want to change the configurations or try other endpoint or chains, you can check the API documentation here: 
- https://www.covalenthq.com/docs/api/#/0/0/USD/1 
 
list of supported networks in here: 
 - https://www.covalenthq.com/docs/networks/ 

```python
import pandas as pd 
import requests


address_list = ['']  # list of addreess you want to get the historical transactional data
api_key = ''  # your covalent API_KEy like 'ckey_0ec0432792...'

session = requests.Session() 
session.auth = (api_key, '')  # you don't need a password 
df = []
df = pd.DataFrame(df)
chain_id = '1' # 1 is the chain_id for ethereum main net 

# URl is configured to not return decoded event logs

for address in address_list: 
  
  response = session.get(f'https://api.covalenthq.com/v1/{chain_id}/address/{address_list}/transactions_v2/?quote-currency=USD&format=JSON&block-signed-at-asc=false&no-logs=true')
  trx = pd.DataFrame(data = response.json()['data']['items'])
  df = pd.concat([df,trx])

df = df.reset_index(drop= True)
df

## example of output:
```
